# Step 2: Conditional Statements
## Students Grading Logic

```python
marks = int(input("Enter your marks: "))
if marks >= 85:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
elif marks >= 65:
    print("Grade: C")
elif marks >= 50:
    print("Grade: D")
else:
    print("Grade: F")
```

# Step 3: Loops and Pattern Printing
## Right Angle Triangle

```python
n = 6
for i in range(1, n+1):
    print("*" * i)
```

## Reverse Right Angle Triangle

```python
n = 6
for i in range(n, 0, -1):
    print("*" * i)
```

## Triangle

```python
n = 6 
for i in range(1, n+1):
    print(" " * (n-i) + "*" * (2*i-1))
```

```python
marks = {
    "Ali": 85,
    "Ahmed": 78,
    "Sara": 92,
    "Fatima": 88,
    "Usman": 81
}

print("Students:", marks.keys())
print("Marks:", marks.values())
print("Highest:", max(marks.values()))
print("Lowest:", min(marks.values()))
print("Average:", sum(marks.values()) / len(marks))
```

# Step 6: Word Frequency Counter

```python
text = "Python is simple yet powerful. Python is easy to learn."
words = text.split()
frequency = {}
for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1
print(frequency)
```

## Character Frequency Counter

```python
word = "Python is simple yet powerful. Python is easy to learn."
freq = {}

for ch in word:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

print(freq)
```

# Step 7: List Mutability in Functions

```python
def add_item(items):
    items.append("Apple")

fruits = ["Banana", "Orange"]

add_item(fruits)

print(fruits)
```

```python
def update(numbers):
    numbers[0] = 100

nums = [10, 20, 30]

update(nums)

print(nums)
```
