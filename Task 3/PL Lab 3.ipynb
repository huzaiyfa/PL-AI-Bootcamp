### Import the required libraries
We will use pandas for data loading and cleaning, and NumPy for matrix-based operations.

```python
# Import the libraries needed for data analysis and numerical operations
import pandas as pd
import numpy as np
```

### Step 1: Load the dataset
We read the CSV file into a DataFrame so we can explore the weather data in a tabular format.

```python
# Load the weather dataset into a pandas DataFrame
# This file contains hourly weather measurements for 10 days

df = pd.read_csv("hourly_weather_10_days.csv")
df
```

### Step 2: Basic Exploration

```python
# Check the column names, non-null counts, and data types
df.info()
```

```python
# Get basic statistical summaries for the numeric columns
df.describe()
```

```python
# Count missing values in each column to see where cleaning is needed
df.isna().sum()
```

### Step 3: Handle missing values

```python
# Fill missing numeric values with the mean of each column
numerical_columns = df.select_dtypes(include=['number']).columns

for column in numerical_columns:
    df[column] = df[column].fillna(df[column].mean())

print(df.isna().sum())
# Confirm that no missing values remain after filling
```

### Step 4: Data Analysis

```python
# Convert the timestamp column into datetime format
# Then extract the date and hour for later grouping

df['timestamp'] = pd.to_datetime(df['timestamp'])
df['date'] = df['timestamp'].dt.date
df['hour'] = df['timestamp'].dt.hour
```

```python
# Compute the average temperature for each day
# Group by date and calculate the mean temperature


daily_avg_temp = df.groupby('date')['temperature_C'].mean()
print('Daily average temperature:')
print(daily_avg_temp)
```

```python
# Calculate min, max, and mean values for the main weather metrics
summary_stats = df[['temperature_C', 'humidity_%', 'wind_speed_kmph', 'pressure_hPa', 'visibility_km']].agg(['min', 'max', 'mean'])
print('\nSummary statistics for each metric:')
print(summary_stats)
```

```python
# Find which hour of the day has the highest average humidity
avg_humidity_by_hour = df.groupby('hour')['humidity_%'].mean()
most_humid_hour = avg_humidity_by_hour.idxmax()
print('\nAverage humidity by hour:')
print(avg_humidity_by_hour)
print(f'\nThe most humid hour on average is {most_humid_hour}:00 with {avg_humidity_by_hour.loc[most_humid_hour]:.2f}% humidity.')
```

### Step 5: Numpy Matrix Excercises

```python
# Convert the selected columns into NumPy arrays for matrix operations
# This is useful for reshaping and other operations that require array manipulation

temp = df['temperature_C'].to_numpy()
wind = df['wind_speed_kmph'].to_numpy()

print('Wind speed array:', wind[:5])  # first five elements of the wind speed array
print('\nTemperature array:', temp[:5])  # first five elements of the temperature array
```

### a) Reshape into Matrix Form

```python
# Reshape the temperature data into a 10x24 matrix
# Each row represents one day, and each column represents one hour

temp_matrix = temp.reshape((10, 24))

# Daily min, max, and mean using axis-based operations
# axis=1 means compute across the 24 hourly values for each day

daily_min = temp_matrix.min(axis=1)
daily_max = temp_matrix.max(axis=1)
daily_mean = temp_matrix.mean(axis=1)

print('Daily mean temperatures:', daily_mean)

print('Temperature matrix shape:', temp_matrix.shape)
print('Daily maximum temperatures:', daily_max)
print('Daily minimum temperatures:', daily_min)
```

### b) Normalize the Temperature Matrix

```python
# Normalize the temperature matrix manually using NumPy
# This subtracts the overall mean and divides by the standard deviation

def normalize(matrix):
    mean = matrix.mean()

    std = matrix.std()    
    return (matrix - mean) / std
```

```python
# Apply the normalization function to the temperature matrix
normalized_temp = normalize(temp_matrix)

print('Mean after normalization:', normalized_temp.mean())
print('Standard deviation after normalization:', normalized_temp.std())
print(normalized_temp)
```

### c) Apply Custom/Mask Filter

```python
# Create a boolean mask for wind speeds greater than 15 km/h
# Then extract only those values from the array
mask = wind > 15
high_wind = wind[mask]

print('High-wind readings:', high_wind)
print('Count of high-wind readings:', high_wind.size)
```

## Final Challenge: Daily Temperature Summary

```python
# Define a custom function to summarize each day of the matrix
# It returns a list of dictionaries with min, max, and mean values

def daily_summary(matrix):
    summaries = []
    for day_index, day_values in enumerate(matrix):
        summaries.append({
            'day': day_index + 1,
            'min': float(day_values.min()),
            'max': float(day_values.max()),
            'mean': float(day_values.mean())
        })
    return summaries

summaries = daily_summary(temp_matrix)    
for summary in summaries:
    print(summary)
```
